```python
# ================================================================
# GHANA BANKING SECTOR PERFORMANCE ANALYSIS (2015-2023)
# Source: Bank of Ghana Annual Reports
# Author: Richard Courage Cobbinah | SetorCourage
# ================================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# ----------------------------------------------------------------
# PLOT STYLING
# ----------------------------------------------------------------
plt.rcParams.update({
    'figure.figsize': (14, 7),
    'figure.dpi': 100,
    'font.size': 12,
    'font.family': 'sans-serif',
    'axes.titlesize': 16,
    'axes.titleweight': 'bold',
    'axes.labelsize': 13,
    'axes.spines.top': False,
    'axes.spines.right': False,
    'xtick.labelsize': 11,
    'ytick.labelsize': 11,
    'legend.fontsize': 11,
    'figure.facecolor': 'white',
    'axes.facecolor': 'white'
})

print("=" * 55)
print(" GHANA BANKING SECTOR PERFORMANCE ANALYSIS")
print(" Period: 2015 - 2023")
print(" Source: Bank of Ghana Annual Reports")
print("=" * 55)
print("Libraries loaded successfully.")
```

```python
# ----------------------------------------------------------------
# SECTION 2: DATA - BANK OF GHANA ANNUAL REPORTS
# ----------------------------------------------------------------

data = {
    'Year': [2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023],
    
    # Total assets (GHS Billion)
    'Total_Assets_GHS_Bn': [71.1, 89.4, 99.3, 105.2, 133.8, 167.4, 210.6, 249.1, 298.4],
    
    # Total deposits (GHS Billion)
    'Total_Deposits_GHS_Bn': [45.2, 57.3, 63.1, 71.4, 93.5, 118.6, 152.3, 178.9, 214.7],
    
    # Total loans and advances (GHS Billion)
    'Total_Loans_GHS_Bn': [28.4, 33.6, 35.2, 37.8, 46.3, 55.1, 68.4, 84.2, 98.6],
    
    # Non-performing loan ratio (%)
    'NPL_Ratio_Pct': [14.7, 17.3, 21.6, 18.2, 14.3, 14.8, 17.3, 20.1, 18.9],
    
    # Capital adequacy ratio (%)
    'CAR_Pct': [17.8, 17.4, 11.8, 19.1, 21.4, 20.2, 19.8, 16.3, 14.7],
    
    # Return on equity (%)
    'ROE_Pct': [22.1, 20.3, 8.4, 15.6, 18.2, 17.9, 19.4, 12.3, 14.8],
    
    # Return on assets (%)
    'ROA_Pct': [3.2, 2.8, 0.9, 2.1, 2.6, 2.5, 2.8, 1.6, 1.9],
    
    # Number of licensed banks
    'Num_Banks': [33, 35, 34, 23, 23, 23, 23, 23, 24]
}

df = pd.DataFrame(data)

print("Dataset created successfully.")
print(f"Shape: {df.shape}")
print("\nPreview:")
print(df.to_string(index=False))
```

```python
# ----------------------------------------------------------------
# SECTION 3: CHART 1 - TOTAL ASSETS, DEPOSITS AND LOANS (2015-2023)
# ----------------------------------------------------------------

fig, ax = plt.subplots(figsize=(14, 7))

ax.plot(df['Year'], df['Total_Assets_GHS_Bn'], 
        marker='o', linewidth=2.5, color='#1a5276', label='Total Assets')
ax.plot(df['Year'], df['Total_Deposits_GHS_Bn'], 
        marker='s', linewidth=2.5, color='#1e8449', label='Total Deposits')
ax.plot(df['Year'], df['Total_Loans_GHS_Bn'], 
        marker='^', linewidth=2.5, color='#b7950b', label='Total Loans & Advances')

ax.set_title('Ghana Banking Sector: Assets, Deposits and Loans (2015-2023)', 
             pad=20)
ax.set_xlabel('Year')
ax.set_ylabel('GHS Billion')
ax.set_xticks(df['Year'])
ax.legend()
ax.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.savefig('chart1_assets_deposits_loans.png', dpi=150, bbox_inches='tight')
plt.show()

print("Chart 1 saved.")
```

```python
# ----------------------------------------------------------------
# SECTION 4: CHART 2 - NPL RATIO AND BANKING SECTOR CLEANUP
# ----------------------------------------------------------------

fig, ax1 = plt.subplots(figsize=(14, 7))

# NPL ratio line
ax1.plot(df['Year'], df['NPL_Ratio_Pct'],
         marker='o', linewidth=2.5, color='#c0392b', label='NPL Ratio (%)')
ax1.set_xlabel('Year')
ax1.set_ylabel('NPL Ratio (%)', color='#c0392b')
ax1.tick_params(axis='y', labelcolor='#c0392b')
ax1.set_xticks(df['Year'])

# Shade the crisis period
ax1.axvspan(2017, 2018, alpha=0.15, color='red', label='Banking Cleanup Period')

# Number of banks on secondary axis
ax2 = ax1.twinx()
ax2.bar(df['Year'], df['Num_Banks'],
        alpha=0.3, color='#2980b9', label='Number of Banks')
ax2.set_ylabel('Number of Licensed Banks', color='#2980b9')
ax2.tick_params(axis='y', labelcolor='#2980b9')

# Annotations
ax1.annotate('NPL peaks at 21.6%\nBank of Ghana begins cleanup',
             xy=(2017, 21.6), xytext=(2015.5, 23),
             arrowprops=dict(arrowstyle='->', color='black'),
             fontsize=10)

ax1.annotate('34 banks reduced\nto 23 after cleanup',
             xy=(2018, 18.2), xytext=(2019.5, 22),
             arrowprops=dict(arrowstyle='->', color='black'),
             fontsize=10)

ax1.set_title('Ghana NPL Ratio and Banking Sector Cleanup (2015-2023)', pad=20)

lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2, loc='upper right')

plt.tight_layout()
plt.savefig('chart2_npl_cleanup.png', dpi=150, bbox_inches='tight')
plt.show()

print("Chart 2 saved.")
```

```python
# ----------------------------------------------------------------
# SECTION 5: CHART 3 - RETURN ON EQUITY AND RETURN ON ASSETS
# ----------------------------------------------------------------

fig, ax = plt.subplots(figsize=(14, 7))

ax.plot(df['Year'], df['ROE_Pct'],
        marker='o', linewidth=2.5, color='#1a5276', label='Return on Equity (%)')
ax.plot(df['Year'], df['ROA_Pct'],
        marker='s', linewidth=2.5, color='#1e8449', label='Return on Assets (%)')

# Shade cleanup period
ax.axvspan(2017, 2018, alpha=0.15, color='red', label='Banking Cleanup Period')

# Annotations
ax.annotate('ROE collapses to 8.4%\nduring cleanup',
            xy=(2017, 8.4), xytext=(2015.5, 6),
            arrowprops=dict(arrowstyle='->', color='black'),
            fontsize=10)

ax.annotate('ROA drops to 0.9%\nin 2017',
            xy=(2017, 0.9), xytext=(2019, 0.5),
            arrowprops=dict(arrowstyle='->', color='black'),
            fontsize=10)

ax.set_title('Ghana Banking Sector: Profitability Ratios (2015-2023)', pad=20)
ax.set_xlabel('Year')
ax.set_ylabel('Percentage (%)')
ax.set_xticks(df['Year'])
ax.legend()
ax.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.savefig('chart3_profitability.png', dpi=150, bbox_inches='tight')
plt.show()

print("Chart 3 saved.")
```

```python
# ----------------------------------------------------------------
# SECTION 6: CHART 4 - CAPITAL ADEQUACY RATIO (2015-2023)
# ----------------------------------------------------------------

fig, ax = plt.subplots(figsize=(14, 7))

bars = ax.bar(df['Year'], df['CAR_Pct'],
              color=['#c0392b' if x < 13 else '#1a5276' for x in df['CAR_Pct']],
              alpha=0.85, width=0.6)

# Regulatory minimum line
ax.axhline(y=13, color='#e74c3c', linewidth=2,
           linestyle='--', label='Regulatory Minimum (13%)')

# Shade cleanup period
ax.axvspan(2016.7, 2017.3, alpha=0.15, color='red')

# Value labels on bars
for bar, val in zip(bars, df['CAR_Pct']):
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.3,
            f'{val}%', ha='center', va='bottom', fontsize=10, fontweight='bold')

ax.annotate('CAR drops to 11.8%\nbelow regulatory minimum',
            xy=(2017, 11.8), xytext=(2019, 10),
            arrowprops=dict(arrowstyle='->', color='black'),
            fontsize=10)

ax.set_title('Ghana Banking Sector: Capital Adequacy Ratio (2015-2023)', pad=20)
ax.set_xlabel('Year')
ax.set_ylabel('Capital Adequacy Ratio (%)')
ax.set_xticks(df['Year'])
ax.set_ylim(0, 28)
ax.legend()
ax.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.savefig('chart4_capital_adequacy.png', dpi=150, bbox_inches='tight')
plt.show()

print("Chart 4 saved.")
```
